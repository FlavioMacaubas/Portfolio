```python
# Leitura dos dados
import pandas as pd

df = pd.read_csv('dim_dtb_rgapb.csv', sep = '|', encoding = 'utf-8')

# Ordenando por gerencia de ensino
df = df.sort_values(by=['_id'], ascending = True)

# Visualizando as 14 observações
df
```

## Ajustando gerências

```python
cod_muni_ordenado = [list(eval(e).values())[0] for e in df.municipios]
```

```python
# Cria lista de escolas por gerencia regional
geren_regio = {}
cod_muni_ordenado = [list(eval(e).values())[0] for e in df.municipios]

for i in df._id.values:
    geren_regio.update({i:cod_muni_ordenado[i-1]})
```

```python
gerencias = pd.read_csv('COesc_Muni.csv', sep = ';')

# Nova coluna de gerencia regional
coluna_ger = []

for linha in gerencias.CO_MUNICIPIO_ESC.values:
    for c,v in geren_regio.items():
        if linha in v:
            coluna_ger.append(c)


# Adicionando nova coluna de gerencia regional
gerencias['GER_REGIONAL'] = coluna_ger

gerencias = gerencias[['CO_ESCOLA', 'GER_REGIONAL']]

# Salvando arquivo
gerencias.to_excel('GER_REGIONAL.xls', index = False)
```
